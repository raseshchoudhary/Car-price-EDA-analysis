# Wheel base vs car length

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['wheelbase'],df['carlength'])
plt.xlabel('Wheel base')
plt.ylabel('Car length')
```

# Interpretations:
>There is a positive correlation berween wheelbase and carlength which means that cars with more length require a broader wheel base.

>Thus if the aim is to make longer cars that the owner should keep in mind that the wheel base is also made broader and vice versa

# Which body type is the costliest?

```python
df.groupby('carbody').mean()['price'].plot(kind='line',color=('red'))
plt.xlabel('Car Body Type')
plt.ylabel('Mean price')
```

# Interpretations
As we saw in the above chart that sadan and hatchback carbody cars are more prefers, well the reason can be the price as we can see that these two car bodies are dead cheap. This indicates that the owner is dealing with price elastic customers.

# Car price vs horsepower

```python
plt.scatter(df['price'],df['horsepower'],color=('green'))
plt.xlabel('Car Price')
plt.ylabel('Horsepower')
```

# Interpretations
There is a positive coorelation between price of the vehicle and its horsepower. Cars with higher horsepower means that they are comparatively stronger with the cars with weaker horsepower and since they are stronger, they need to be made with sturdier equiments that will eventuall cost more.

# Relation between engine size and car price

```python
plt.scatter(df['price'],df['enginesize'],color=('red'))
plt.xlabel('Car price')
plt.ylabel('Engine size')
```

# Interpretation
The more the engine size, the costlier the vehicle. If we combine the the above two graphs we can get to the conclusion that
a large engine size leads to more horse power, in short engine size and horse power are also correlated. (Refer to the histogram below for more elaboration).

# Distribution of engine size and horsepower

```python
df['enginesize'].hist(bins=20)
df['horsepower'].hist(bins=20)
plt.ylabel('Count')
```

# Interpretation
As we can see that the histograms of both the engine size and horsepower are almost alike which indicates that one effects the other in a positive manner. This is to make the customer understand that if they wish to have a sturdy car with a high horse power that they need a car with a bigger engine which will eventually cost more. In other words engine size, horsepower and price of the car are mutually correlated with each other on a positive manner as indicated by the graphs above as well as by the histogram.
